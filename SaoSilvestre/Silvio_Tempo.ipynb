```python
import pandas as pd
```

```python
dt = pd.read_csv("Analise.csv", encoding='latin-1')
dt.head()
```

```python
dt[['hour_TempoM', 'minute_TempoM', 'second_TempoM']] = dt['TempoM'].astype(str).str.split(':', expand=True).astype(int)
dt[['hour_TempoH', 'minute_TempoH', 'second_TempoH']] = dt['TempoH'].astype(str).str.split(':', expand=True).astype(int)
```

```python
dt.drop('second_TempoM', axis=1, inplace=True)
dt.drop('second_TempoH', axis=1, inplace=True)
dt= dt.rename(columns={'hour_TempoM': 'minM', 'minute_TempoM': 'segM','hour_TempoH':'minH','minute_TempoH':'segH'} )
dt.head()
```
